```python
import glob

suw_csv_paths = glob.glob('/autofs/diamond2/share/corpus/CEJC2304/data/**/*SUW.csv', recursive=True)
suw_csv_paths.sort()
suw_csv_paths[:5]  # 先頭5件だけ表示（確認用）
```

```python
import pandas as pd

for path in suw_csv_paths:
    df = pd.read_csv(path, encoding='Shift_JIS')
    pron_list = df['発音'].tolist()
    pron_list = [pron for pron in pron_list if '(' in pron or ')' in pron]
    if pron_list:
        print(pron_list)

        
```

```python
# LUWでも同様のことをする
luw_csv_paths = glob.glob('/autofs/diamond2/share/corpus/CEJC2304/data/**/*LUW.csv', recursive=True)
luw_csv_paths.sort()
for path in luw_csv_paths:
    df = pd.read_csv(path, encoding='Shift_JIS')
    pron_list = df['発音'].tolist()
    pron_list = [pron for pron in pron_list if '(' in pron or ')' in pron]
    if pron_list:
        print(pron_list)
```

```python
# '発音'列に'('が含まれる行を抽出して表示
for path in suw_csv_paths:
    df = pd.read_csv(path, encoding='Shift_JIS')
    mask = df['発音'].str.contains('\(', na=False)
    extracted_rows = df[mask]    
    if not extracted_rows.empty:
        print(f"File: {path}")
        display(extracted_rows)
        print("\n")  # 改行で区切る
```
